```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.cluster import DBSCAN
import matplotlib.pyplot as plt
import numpy as np
```

```python
hud_awards = pd.read_csv("../Resources/Clean Data/HUD_Award_2005-2023_clean.csv")
hud_awards.head()
```

```python
ovho_beds_df = pd.read_csv("../Resources/Clean Data/OvHo_beds.csv")
ovho_beds_df.head()
```

```python
hud_award_by_state = hud_awards.groupby('State')['Award Amount'].mean().reset_index()

ovho_beds_aggregated = ovho_beds_df.groupby('State').agg({'Overall Homeless': 'mean', 'Total Beds ES, TH, PSH': 'mean'}).reset_index()

state_level_merged = pd.merge(ovho_beds_aggregated, hud_award_by_state, on='State', how='inner')

scaler = StandardScaler()
X_scaled = scaler.fit_transform(state_level_merged[['Overall Homeless', 'Total Beds ES, TH, PSH', 'Award Amount']])
```

```python
inertia = []
for n_clusters in range(1, 11): 
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    kmeans.fit(X_scaled)
    inertia.append(kmeans.inertia_)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(range(1, 11), inertia, marker='o')
plt.title('Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.xticks(range(1, 11)) 
plt.grid(True)
plt.show()
```

```python
kmeans = KMeans(n_clusters=3, random_state=42)
state_level_merged['Cluster'] = kmeans.fit_predict(X_scaled)
clusters = kmeans.labels_
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

colors = ['blue', 'red', 'green']

plt.figure(figsize=(12, 8))

for cluster_index, color in enumerate(colors):
    cluster_data = X_pca[state_level_merged['Cluster'] == cluster_index]
    
    plt.scatter(cluster_data[:, 0], cluster_data[:, 1], color=color, label=f'Cluster {cluster_index}')

for index, row in state_level_merged.iterrows():
    if row['Cluster'] in [1, 2]:  
        plt.text(X_pca[index, 0], X_pca[index, 1], row['State'], fontsize=12, color='black', ha='right', va='bottom')

plt.title('PCA of State-Level Data')
plt.xlabel('Primary Variance In Homelessness Metrics and Funding')
plt.ylabel('Secondary Variance In Homelessness Metrics and Funding')


plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
%matplotlib notebook
from mpl_toolkits.mplot3d import Axes3D
from sklearn.decomposition import PCA

pca_3d = PCA(n_components=3)
X_pca_3d = pca_3d.fit_transform(X_scaled)

fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

colors = ['r', 'g', 'b']

for i in range(3):
    ax.scatter(X_pca_3d[clusters == i, 0], X_pca_3d[clusters == i, 1], X_pca_3d[clusters == i, 2], 
               c=colors[i], label=f'Cluster {i}', marker='o')

ax.set_xlabel('Principal Component 1')
ax.set_ylabel('Principal Component 2')
ax.set_zlabel('Principal Component 3')
ax.set_title('3D PCA Plot')
plt.show()

```

```python
feature_names = ['Overall Homeless', 'Total Beds ES, TH, PSH', 'Award Amount']  

pca_3d_components = pca_3d.components_

components_df_3d = pd.DataFrame(pca_3d_components, 
                             columns=feature_names, 
                             index=[f'PC{i+1}' for i in range(pca_3d_components.shape[0])])

components_df_3d
```
